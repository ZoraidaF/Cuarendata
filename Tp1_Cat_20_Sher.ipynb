```python
fig1longtext = px.parallel_coordinates(long_target, color="longText", labels={'Verdadero','longText','Falso',},
                             color_continuous_scale=[(0.00, "red"),   (0.33, "red"),
                                                     (0.33, "green"), (0.66, "green"),
                                                     (0.66, "blue"),  (1.00, "blue")])

fig1longtext.show()
fig1longtext.write_html("longtext.html")
```
